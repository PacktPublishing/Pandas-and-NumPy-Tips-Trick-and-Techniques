# Pandas Concanenate

```python
import pandas as pd
import pandas_datareader as pdr
```

### Concatenation is essentially adding either columns or rows to a dataframe

```python
aapl = pdr.get_data_yahoo('AAPL', '20190901')
goog = pdr.get_data_yahoo('GOOG', '20190901')
```

```python
pd.concat([aapl, goog]).tail()
```

```python
astr = ['AAPL'] *len(aapl)
aapl.set_index([astr, aapl.index], inplace=True)
```

```python
gs = ['GOOG'] * len(goog)
goog.set_index([gs,goog.index],inplace=True)
```

```python
aapl.head()
```

```python
combined = pd.concat([aapl,goog])
combined
```

```python
combined.loc['GOOG']
```
