# Join Operations

```python
import pandas as pd
import pandas_datareader as pdr
```

```python
aapl = pdr.get_data_yahoo('AAPL', '20190901')[['Open', 'Close']]
goog = pdr.get_data_yahoo('GOOG', '20190825')[['Open', 'Close']]
```

### Really a shortcut method of .merge exclusively for index merges

```python
aapl.join(goog, rsuffix='_GOOG', lsuffix='_AAPL')
```

```python
aapl.join(goog, how='outer', rsuffix='_GOOG', lsuffix='_AAPL')
```

```python
aapl.join(goog, how='left', rsuffix='_GOOG', lsuffix='_AAPL')
```

### Can be used as a "lookup" to merge an index on column values

```python
var_cost = pd.DataFrame({'cost': [24,23,22,21,20]}, index=[5000,10000, 15000, 20000, 25000])
production = pd.DataFrame({'Price': [40, 41,42,43,44], 'Demand': [25000, 20000,15000,10000,5000]})
```

```python
production.join(var_cost, on='Demand')
```
