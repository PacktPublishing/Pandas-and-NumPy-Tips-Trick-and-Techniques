# Up/Down-sampling
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.resample.html

```python
import pandas as pd
import pandas_datareader as pdr
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
aapl = pdr.get_data_yahoo('AAPL','20180801')
aapl.head()
```

```python
aapl.info()
```

### When resampling you need an aggregation method

```python
aapl.resample('bm').last()
```

### Be sure to validate data

```python
aapl.resample('bms').first()
```

```python
# sampling options = B, M (end), BM(end), MS (month start), BMS
```

### Get a specific day of the week

```python
fridays =aapl.resample('w-fri').last()
fridays.head()
```

### Calculate week over week change

```python
fridays['change'] = fridays['Close'].pct_change()
```

```python
fridays.head()
```

```python
month = aapl.resample('M').mean()
month.head()
```

```python
up_month = month.resample('D').mean().fillna(method='ffill')
up_month.head()
```

```python
up_month['Close'].plot()
```
